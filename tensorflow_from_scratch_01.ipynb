```python
import keras
import numpy as np
```

`keras.Sequential`：這是一種線性堆疊模型，它允許你按順序將多個層加到模型中。每一層都會接收來自前一層的輸出，並將結果傳遞給下一層。

`keras.layers.InputLayer(shape=(1,))`：這層定義了模型的輸入層。在這裡，shape=(1,) 表示輸入的數據將有一個特徵，也就是每個樣本的輸入是一個一維的數據。

`keras.layers.Dense(1)`：這是一個全連接層（Dense layer），它有一個神經元。這層的輸出將是一個標量值。

```python
model = keras.Sequential([
    keras.layers.InputLayer(shape=(1,)),
    keras.layers.Dense(1)
])
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
model.summary()
```

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
```

```python
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```python
model.fit(xs, ys, epochs=1000)
```

```python
print(model.predict(np.array([12.0])))
```
